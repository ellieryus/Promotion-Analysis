```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt

DATA_PATH = '/Users/ellieryus/Downloads/Year4-1/Current Topic/data/'

file_name = DATA_PATH + 'FactSales.csv'

FactSales_df = pd.read_csv(file_name)
FactSales_df.head()
```

```python
file_name2 = DATA_PATH + 'DimTerritory.csv'

DimTerritory_df = pd.read_csv(file_name2)
DimTerritory_df.head()
```

```python
table_FactSales_Territory = pd.merge(DimTerritory_df, FactSales_df)
table_FactSales_Territory.head()
```

```python
table_FactSales_Territory_gb = table_FactSales_Territory.groupby('TerritoryID', as_index=False)
type(table_FactSales_Territory_gb)
table_FactSales_Territory_df = table_FactSales_Territory_gb.describe()
table_FactSales_Territory_df
```

```python
file_name3 = DATA_PATH + 'DimDate.csv'

DimDate_df = pd.read_csv(file_name3)
DimDate_df
```

```python
table_Territory_Date = pd.merge(DimDate_df, table_FactSales_Territory)
table_Territory_Date
```

```python
months_to_filter = [6, 7, 8]
table_Territory_Date_filter = table_Territory_Date[table_Territory_Date['Month'].isin(months_to_filter)]
table_Territory_Date_filter
```

```python
table_Territory_Date_filter = table_Territory_Date_filter[table_Territory_Date_filter.SpecialOfferID != 1]
table_Territory_Date_filter
```

```python
table_Territory_Date_filter_duplicate = table_Territory_Date_filter.drop_duplicates(subset=['SalesOrderID'])
table_Territory_Date_filter_duplicate
```

```python
table_Territory_Date_gb = table_Territory_Date_filter_duplicate.groupby('TerritoryID', as_index=False).sum(['TotalDue', 'OrderQty'])

table_Territory_Date_gb
```

```python
table_df = pd.DataFrame()
table_df['TerritoryID'] = table_Territory_Date_gb['TerritoryID']
table_df['State'] = ['Oregon, Washington, Idaho, Montana, Wyoming', 
                    'Maine, New York, New Jersey, Vermont, Massachusetts, Rhode Island, Connecticut, New Hampshire, Pennsylvania', 
                    'Illinois, Indiana, Iowa, Kansas, Michigan, Minnesota, Missouri, Nebraska, North Dakota, Ohio, South Dakota, Wisconsin', 
                    'Arizona, New Mexico, Nevada, Utah', 
                    'Alabama, Arkansas, Florida, Georgia, Kentucky, Louisiana, Mississippi, North Carolina, South Carolina, Tennessee, Virginia, West Virginia', 
                    'Canada', 
                    'France', 
                    'Germany', 
                    'Australia', 
                    'United Kingdom']
table_df['TotalOrderQty'] = table_Territory_Date_gb['OrderQty']
table_df['TotalDue'] = table_Territory_Date_gb['TotalDue']
table_df

table_df = table_df.merge(table_Territory_Date_filter[['TerritoryID', 'Name']], on='TerritoryID', how='left')

table_df_grouped = table_df.groupby(['TerritoryID', 'Name', 'State']). agg({'TotalOrderQty': 'sum', 'TotalDue': 'sum'})

table_df_grouped
```

```python
table_df_grouped_reset = table_df_grouped.reset_index()

table_df_grouped_reset.to_csv('SalesOfEachTerritorySummer.csv', index=False)
```
