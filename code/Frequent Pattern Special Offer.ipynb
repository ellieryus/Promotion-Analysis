```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import mlxtend
from mlxtend.frequent_patterns import apriori, association_rules 

# Please specify the data path of the files you downloaded
DATA_PATH = '/Users/ellieryus/Downloads/Year4-1/Current Topic/data/'

file_name = DATA_PATH + 'FactSales.csv'
FactSales_df = pd.read_csv(file_name)
FactSales_df
```

```python
FactSales_filter = FactSales_df[FactSales_df['SpecialOfferID'] != 1]
FactSales_filter
```

```python
FactSales_filter_dropduplicate = FactSales_filter.drop_duplicates(subset=['SalesOrderID','SpecialOfferID'])
FactSales_filter_dropduplicate
```

```python
table = FactSales_filter_dropduplicate.groupby(['SalesOrderID']).agg({'SpecialOfferID': list, 'TerritoryID': np.mean})
table.head(100)
```

```python
table['TerritoryID'] = table['TerritoryID'].astype(int)
table
```

```python
table2 = table.groupby(['TerritoryID']).agg(list)
table2
```

```python
from mlxtend.preprocessing import TransactionEncoder

table_Territory_1 = table2.loc[1]['SpecialOfferID']
print(table_Territory_1)
type(table_Territory_1)
```

```python
te = TransactionEncoder()
te_ary = te.fit(table_Territory_1).transform(table_Territory_1)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori

apriori(df, min_support=0.001, use_colnames=True)
```
