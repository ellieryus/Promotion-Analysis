```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import random
import json
import os

# Please specify the data path of the files you downloaded
DATA_PATH = '/Users/ellieryus/Downloads/Year4-1/Current Topic/data/'

file_name = DATA_PATH + 'DimSpecialOffer.csv'

DimSpecialOffer_df = pd.read_csv(file_name)
DimSpecialOffer_df
```

```python
file_name2 = DATA_PATH + 'FactSales.csv'

FactSales_df = pd.read_csv(file_name2)
FactSales_df
```

```python
file_name3 = DATA_PATH + 'DimDate.csv'

DimDate_df = pd.read_csv(file_name3)
DimDate_df
```

```python
file_name4 = DATA_PATH + 'DimProduct.csv'
DimProduct_df = pd.read_csv(file_name4)
DimProduct_df
```

```python
file_name5 = DATA_PATH + 'DimTerritory.csv'
DimTerritory_df = pd.read_csv(file_name5)
DimTerritory_df
```

```python
table = FactSales_df.merge(DimSpecialOffer_df).merge(DimDate_df).merge(DimProduct_df, on='ProductID')
table = table[table.SpecialOfferID != 1]
table
```

```python
table_gb = table.groupby(['SpecialOfferID', 'Year', 'TerritoryID', 'ProductID'], as_index=False).sum(['LineTotal', 'OrderQty'])
table_gb
```

```python

table_df = pd.DataFrame()
table_df['SpecialOfferID'] = table_gb['SpecialOfferID']
table_df['ProductID'] = table_gb['ProductID']
table_df['TotalOrderQuantity'] = table_gb['OrderQty'].astype(int)
table_df['TerritoryID'] = table_gb['TerritoryID']
table_df['Year'] = table_gb['Year']
table_df['LineTotal'] = table_gb['LineTotal']

table_df = table_df.merge(table[['SpecialOfferID', 'Description']], on='SpecialOfferID', how='left')
table_df = table_df.merge(table[['ProductID', 'Name']], on='ProductID', how='left')

table_df_grouped = table_df.groupby(['SpecialOfferID', 'Description', 'ProductID', 'Name', 'Year', 'TerritoryID']).agg({'TotalOrderQuantity': 'sum', 'LineTotal': 'sum'})

table_df_grouped
```

```python

table_df_grouped_reset = table_df_grouped.reset_index()

table_df_grouped_reset.to_csv('SpecialOfferSalesProduct.csv', index=False)
```
